```python
import cv2
```

```python
# cv.imread(path, flag)
# path: path of image directory
# flag: default value in cv2.IMREAD_COLOR

path = r"""F:\OpenCV  Tutorials\Image_data\cat03.jpg"""

# Reading the file from directory
img = cv2.imread(path)    # Press Shifit + tab to get any function detail

# To Show Colored images
img = cv2.imread(path, cv2.IMREAD_COLOR) 
img = cv2.imread(path, 1)  # we can write 1 to show colored image

print(img.shape)
print(img)   # Prints pixels of image as an array. Pixel range from 0 to 255

# Showing the image in Window

cv2.imshow("Image", img)

cv2.waitKey(0)

cv2.destroyAllWindows()


```
