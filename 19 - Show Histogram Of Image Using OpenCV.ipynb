# Histogram Function Explaination

plt.hist(img.ravel(), bins=256, range=[0, 255]): This line of code creates a histogram of pixel intensity values. Here's what each parameter does:

img.ravel(): The ravel() function is used to flatten the 2D (or multi-dimensional) image array into a 1D array. This is necessary for creating a histogram because histograms are typically constructed from a 1D list of values.

bins=256: This parameter specifies the number of bins or divisions in the histogram. In this case, it's set to 256, which means the histogram will have 256 bars, each representing a range of pixel intensity values.

range=[0, 255]: This parameter defines the range of values for the x-axis of the histogram. In this case, it's set from 0 to 255, which covers the entire range of pixel intensity values in an 8-bit image (where pixel values usually range from 0 to 255).

plt.show(): This line of code displays the histogram plot. Without this line, the plot would be created but not shown.

```python
# Histogram of image

plt.hist(img.ravel(), bins = 256, range = [0,255])
plt.show()
```

# Histogram of each channel - Function Explaination
colors = ('b', 'g', 'r'): This line defines a tuple named colors containing three elements, each representing a color channel: 'b' for Blue, 'g' for Green, and 'r' for Red.

img_ravel = [img[:,:,0].ravel(), img[:,:,1].ravel(), img[:,:,2].ravel()]: This line creates a list named img_ravel containing three flattened arrays, each representing the pixel intensity values for one color channel. Here's what each part does:

img[:,:,0].ravel(): This extracts the pixel values from the Red channel of the image ([:,:,0] selects the first channel) and then flattens it into a 1D array using .ravel().

img[:,:,1].ravel(): This does the same for the Green channel.

img[:,:,2].ravel(): This does the same for the Blue channel.

plt.hist(img_ravel, color=colors, label=colors): This line creates a histogram plot using the plt.hist() function. Here's what each parameter does:

img_ravel: This is the list of flattened arrays representing pixel values for the three color channels. The histogram will be plotted for each of these arrays.

color=colors: This parameter assigns the colors defined earlier to each histogram plot. The first histogram will be plotted in blue, the second in green, and the third in red.

label=colors: This adds labels to the histograms, using the color names as labels.

plt.legend(): This line adds a legend to the plot, indicating which color corresponds to which color channel.

plt.show(): This line displays the histogram plot. Without this line, the plot would be created but not shown.

```python
# Histogram of All Channels
colors = ('b','g','r')

img_ravel = [img[:,:,0].ravel(), img[:,:,1].ravel(), img[:,:,2].ravel()]

plt.hist(img_ravel, color = colors, label = colors)

plt.legend()

plt.show()
```

```python
colors = ('b', 'g', 'r')

img_ravel = [img[:, :, 0].ravel(), img[:, :, 1].ravel(), img[:,:, 2].ravel()] 

plt.hist(img_ravel, color=colors, label=colors, bins=256, range=[0,256])

plt.legend()

plt.show()
```

# Plot all channels using cv2.calcHist() function

colors = ('b', 'g', 'r'): This line defines a tuple named colors containing three elements, each representing a color channel: 'b' for Blue, 'g' for Green, and 'r' for Red.

plt.figure(figsize=(16, 9)): This line creates a new figure for the plot with a specified figure size of 16 units in width and 9 units in height. This is done to ensure a larger and more readable plot.

for i, color in enumerate(colors):: This line starts a loop that iterates over the colors tuple using the enumerate() function. The loop assigns an index i and a corresponding color value from the tuple to each iteration.

Inside the loop:

histogram = cv2.calcHist([img], [i], None, [256], [0, 256]): This line calculates the histogram for the current color channel using the cv2.calcHist() function. Here's what each parameter does:

[img]: A list containing the input image.

[i]: A list containing the index of the current color channel (0 for Blue, 1 for Green, and 2 for Red).

None: The mask, which is set to None here, meaning the entire image will be used to calculate the histogram.

[256]: The number of bins for the histogram, set to 256 to cover the entire range of pixel intensity values.

[0, 256]: The range of pixel intensity values to consider for the histogram.

plt.plot(histogram, color=color): This line plots the histogram calculated for the current color channel using the plt.plot() function. The color parameter is set to the current channel's color.

After the loop, plt.show() is used to display the plot.

```python
# Plot all channels using cv2.calcHist() function

colors = ('b','g','r')

plt.figure(figsize = (16,9))

for i, color in enumerate(colors):
    histogram = cv2.calcHist([img], [i], None, [256], [0,256])
    plt.plot(histogram, color = color)
    
plt.show()

```
