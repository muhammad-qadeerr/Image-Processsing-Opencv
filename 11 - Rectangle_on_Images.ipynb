```python
import cv2
import numpy as np
```

```python
# Put Rectangle of Image

img_path = r"""F:\OpenCV  Tutorials\Image_data\cat13.jpg"""
img = cv2.imread(img_path)
img = cv2.resize(img,(1280,720))

# pt1 and pt2: These variables define the coordinates of two points that will form the opposite corners of the rectangle to be drawn. 
# pt1 corresponds to the top-left corner, and pt2 corresponds to the bottom-right corner of the rectangle.

pt1 = (400,40)
pt2 = (800,300)
color = (0,255,0) # (B,G,R)
thickness = 4
lineType = cv2.LINE_4
# rectangle(img, pt1, pt2, color[, thickness[, lineType[, shift]]])
img_rectangle = cv2.rectangle(img, pt1, pt2, color, thickness, lineType)
cv2.imshow("Rectangle Images", img_rectangle)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Fill Rectangle of image

img_path = r"""F:\OpenCV  Tutorials\Image_data\cat13.jpg"""
img = cv2.imread(img_path)
img = cv2.resize(img,(1280,720))

# pt1 and pt2: These variables define the coordinates of two points that will form the opposite corners of the rectangle to be drawn. 
# pt1 corresponds to the top-left corner, and pt2 corresponds to the bottom-right corner of the rectangle.

pt1 = (400,40)
pt2 = (800,300)
color = (0,255,0) # (B,G,R)
thickness = -1   # Thickness changes to -1 fill the rectangle. 

lineType = cv2.LINE_4
# rectangle(img, pt1, pt2, color[, thickness[, lineType[, shift]]])
img_rectangle = cv2.rectangle(img, pt1, pt2, color, thickness, lineType)
cv2.imshow("Rectangle Images", img_rectangle)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Text Above Rectangle on an Image
img_path = r"""F:\OpenCV  Tutorials\Image_data\cat13.jpg"""
img = cv2.imread(img_path)
img = cv2.resize(img, (1280,720))

pt1 = (400,40)
pt2 = (800,300)
color = (0,255,0)
thickness = 4
lineType = cv2.LINE_4

img_rect = cv2.rectangle(img,pt1,pt2,color,thickness,lineType)

text = "Cat - 100 %"
org = (400, 35)
fontFace = cv2.FONT_HERSHEY_SIMPLEX
fontScale = 1
color = (255, 0, 0)
thickness = 3
lineType = cv2.LINE_4
bottomLeftOrigin = False
# Putting  text on already existing rectangle
# putText(img, text, org, fontFace, fontScale, color[, thickness[, lineType[, bottomLeftOrigin]]]) 
img_rect_text = cv2.putText(img_rect,text,org,fontFace,fontScale,color,thickness,lineType,bottomLeftOrigin)
cv2.imshow("Rectangle Image", img_rect_text)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Text Inside Rectangle on an Image - Just change the origin where you want to put the text
img_path = r"""F:\OpenCV  Tutorials\Image_data\cat13.jpg"""
img = cv2.imread(img_path)
img = cv2.resize(img, (1280,720))

pt1 = (400,40)
pt2 = (800,300)
color = (0,255,0)
thickness = 4
lineType = cv2.LINE_4

img_rect = cv2.rectangle(img,pt1,pt2,color,thickness,lineType)

text = "Cat - 100 %"
org = (500, 170)
fontFace = cv2.FONT_HERSHEY_SIMPLEX
fontScale = 1
color = (255, 0, 0)
thickness = 3
lineType = cv2.LINE_4
bottomLeftOrigin = False
# Putting  text on already existing rectangle
# putText(img, text, org, fontFace, fontScale, color[, thickness[, lineType[, bottomLeftOrigin]]]) 
img_rect_text = cv2.putText(img_rect,text,org,fontFace,fontScale,color,thickness,lineType,bottomLeftOrigin)
cv2.imshow("Rectangle Image", img_rect_text)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
