```python
import cv2
import numpy as np
```

```python
video_path1 = r'F:\OpenCV  Tutorials\Video_data\Street.mp4'
video_path2 = r'F:\OpenCV  Tutorials\Video_data\Cat.mp4'
```

```python
# Open the video file using OpenCV's VideoCapture function
cap = cv2.VideoCapture(video_path1)
# Start a loop to process each frame of the video
while cap.isOpened():
     # Read the current frame from the video
    ret, frame = cap.read()                    #  ret is a boolean indicating whether the read was successful, and frame holds the actual frame data.
    # Check if the frame was read successfully
    if ret:
         # Resize the frame to a specific size (600x600 pixels)
        image = cv2.resize(frame, (1280,720))
        # Show the resized frame in a window named "Video Player"
        cv2.imshow("Video Player: ", image)
        # Wait for a key press for 25 milliseconds, and check if the pressed key is 'q'
        if cv2.waitKey(25) & 0xff == ord('q'):
            # If 'q' is pressed, exit the loop
            break
    else:
         break
# Release the video capture object to free up resources
cap.release()
cv2.destroyAllWindows()  
```

```python
# Capturing or adding multiple videos in a single frame

import cv2
import numpy as np

path = r"""F:\OpenCV Tutorials\Image_data"""
img_name = "cat13.jpg"

video_path1 = r'F:\OpenCV  Tutorials\Video_data\Street.mp4'
video_path2 = r'F:\OpenCV  Tutorials\Video_data\Cat.mp4'

cap1 = cv2.VideoCapture(video_path1)
cap2 = cv2.VideoCapture(video_path2)
cap3 = cv2.VideoCapture(0)

while cap1.isOpened() and cap2.isOpened() and cap3.isOpened():
    ret1, frame1 = cap1.read()
    ret2, frame2 = cap2.read()
    ret3, c_frame = cap3.read()

    if ret1 and ret2 and ret3:
        frame1 = cv2.resize(frame1, (600, 400))
        frame2 = cv2.resize(frame2, (600, 400))

        camera_frame = cv2.resize(c_frame, (frame2.shape[1], frame2.shape[0]))

        image = cv2.imread(path + '\\' + img_name)
        if image is not None:
            image = cv2.resize(image, (600, 400))

        frame_2 = np.hstack((frame1, frame2))
        
        # Resize the camera frame to match the width of frame_2
        camera_frame_resized = cv2.resize(camera_frame, (frame_2.shape[1], frame_2.shape[0]))

        frame_combined = np.vstack((camera_frame_resized, frame_2))

        cv2.imshow("Video Player", frame_combined)

        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    else:
        break

cap1.release()
cap2.release()
cap3.release()
cv2.destroyAllWindows()
```
