```python
# Saving or Converting a random Array into an Image
import cv2
import numpy as np
import os

rand_arr = np.random.randint(255, size = (300,600,3))    # randint(low, high=None, size=None, dtype=int)

cv2.imwrite(r"F:\OpenCV  Tutorials\Image_data\randNpArray.png", rand_arr)   # imwrite(filename, img[, params])

img = cv2.imread(r"F:\OpenCV  Tutorials\Image_data\randNpArray.png")
cv2.imshow('Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Reading and saving image to some other or current directory
import cv2
import numpy as np
import os

img_path = r"""F:\OpenCV  Tutorials\Image_data\cat08.jpg"""
img = cv2.imread(img_path, 0)  # Set the parameter 0 it will read image in grayscale
cv2.imshow("Image", img)

# Saving the image in current directory, though we can specify directory we want to save it in.
cv2.imwrite("cat08_grayscale.png", img)  # saving in the directory as grayscale bcz reading it as grayscale.

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Saving Image from a video

import cv2
import numpy as np
import os
# The r before the triple quotes indicates that the string should be treated as a raw string, which is useful for file paths to avoid issues with backslashes.
video_path = r"""F:\OpenCV  Tutorials\Video_data\CellPhone.mp4"""

os.mkdir("video_to_image")  # Creating a directory to save the video

# Capturing a Video
cap = cv2.VideoCapture(video_path)     # cv2.VideoCapture(self, /, *args, **kwargs)

img_count = 1

while cap.isOpened():
    ret, frame = cap.read() # Reading the frames of video (ret is a boolean that indicates whether a frame was successfully read.)
    
    if not ret:
        print("Unable to read video frames! ")
        
    is_img_write = cv2.imwrite(f"video_to_image\image{img_count}.jpg", frame)
    
    if is_img_write:
        print(f'Image saved to video_to_image\image{img_count}.jpg')
    
    cv2.imshow("Video", frame)
    
    cv2.waitKey(25) & 0xff == ord('q')
    img_count += 1
    
cap.release()
cv2.destroyAllWindows()
    

```
